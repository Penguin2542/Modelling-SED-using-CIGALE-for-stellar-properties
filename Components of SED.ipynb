```python
from astropy.io import fits
from astropy.table import Table
from matplotlib import pyplot as plt
```

```python
data_file = fits.open('C:/Users/Arghya Chaudhuri/out/61019923_best_model.fits') # Open the FITS file
data_file.info() # Print out the information about the FITS file 
# data_file has two HDUs, the second one is the data we are interested in. If we are interested in the first HDU, we would use data_file[0].header or data_file[0].data. The first HDU is the header, the second HDU is the data.
data = data_file[1].data # Get the data in the first HDU    
data_file.close() # Close the FITS file
```

```python
data_table = Table(data) # Convert the data to an astropy table
data_table
```

```python
import numpy as np
epsilon = 1e-10
plt.plot(np.log(data_table['wavelength']), np.log(data_table['stellar.old']+epsilon), 'r-',label="old stellar emission") # Plot the SFR vs age
plt.plot(np.log(data_table['wavelength']), np.log(data_table['stellar.young']+epsilon), 'b-',label="young stellar emission")
plt.xlabel('Wavelength') # Label the x-axis
plt.ylabel('Stellar Emission') # Label the y-axis 
plt.title('Stellar emission vs Wavelength') # Title the plot
#plt.xlim(0,1)
plt.legend()
plt.show() # Show the plot
```
